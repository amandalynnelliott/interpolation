# SciKit Examples
### Noiseless Case

```python
np.random.seed(1)

def f(x):
     return x * np.sin(x)


X = np.atleast_2d([1., 3., 5., 6., 7., 8.]).T
print(X)
```

```python
y = f(X).ravel()
print(y)
```

```python
x = np.atleast_2d(np.linspace(0, 10, 1000)).T
```

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
print(kernel)
```

```python
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
```

```python
gp.fit(X,y)
```

```python
y_pred, sigma = gp.predict(x, return_std = True)
```

```python
plt.figure()
plt.plot(x, f(x), 'r:', label=r'$f(x) = x\,\sin(x)$')
plt.plot(X, y, 'r.', markersize=10, label='Observations')
plt.plot(x, y_pred, 'b-', label='Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-10, 20)
plt.legend(loc='upper left')
```

### Noisy Case

```python
X = np.linspace(0.1, 9.9, 20)
print(X)
```

```python
X = np.atleast_2d(X).T
print(X)
```

```python
y = f(X).ravel()
dy = 0.5 + 1.0 * np.random.random(y.shape)
noise = np.random.normal(0, dy)
y += noise
```

```python
gp = GaussianProcessRegressor(kernel=kernel, alpha=dy ** 2,
                              n_restarts_optimizer=10)

gp.fit(X, y)

y_pred, sigma = gp.predict(x, return_std=True)
```

```python
plt.figure()
plt.plot(x, f(x), 'r:', label=r'$f(x) = x\,\sin(x)$')
plt.errorbar(X.ravel(), y, dy, fmt='r.', markersize=10, label='Observations')
plt.plot(x, y_pred, 'b-', label='Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-10, 20)
plt.legend(loc='upper left')

plt.show()
```

### SciPy Cubic-spine Interpolation example

```python
x = np.arange(0, 2*np.pi+np.pi/4, 2*np.pi/8)
y = np.sin(x)
tck = interpolate.splrep(x, y, s=0)
xnew = np.arange(0, 2*np.pi, np.pi/50)
ynew = interpolate.splev(xnew, tck, der=0)

plt.figure()
plt.plot(x, y, 'x', xnew, ynew, xnew, np.sin(xnew), x, y, 'b')
plt.legend(['Linear', 'Cubic Spline', 'True'])
plt.axis([-0.05, 6.33, -1.05, 1.05])
plt.title('Cubic-spline interpolation')
plt.show()
```

### Applying SciKit to PtPos 171.5 mass

```python
X = []
y = []

events = ptpos['binContent']
data = ptpos['ptpos']

X = np.array(data[:18])
y = np.array(events[:18])

# X = data[:18]
# y = events[:18]

print(np.array(X).shape)
print(np.array(y).shape)

plt.scatter(X, y)
plt.title('PtPos at 171.5 Mass')
plt.xlabel('PtPos')
plt.ylabel('Events')
plt.show()
```

```python
np.random.seed(1)

data = ptpos['ptpos']
binError = ptpos['binError']
binContent = ptpos['binContent']

X = data[:18].to_numpy()
y = binContent[:18].to_numpy()
dy = binError[:18].to_numpy() # noise

X = np.atleast_2d(X).T

# TODO: Try different kernels
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))

# alpha: error at each datapoint
# normalize_y = True  produces better agreement with input template data 
gp = GaussianProcessRegressor(kernel=kernel, alpha=dy,
                              n_restarts_optimizer=10, normalize_y=True)

gp.fit(X, y) 

# Domain to evaluate interpolation at
xmin = ptpos['ptpos'].iloc[0] - ptpos['binWidth2'].iloc[0]
xmax = ptpos['ptpos'].iloc[-1] - ptpos['binWidth2'].iloc[-1]
x = np.atleast_2d(np.linspace(xmin, xmax, 1000)).T

# Do interpolation
y_pred, sigma = gp.predict(x, return_std=True)


plt.figure()

# Known points
plt.errorbar(X.ravel(), y, dy, fmt='r.', markersize=10, label='Observations')

# Interpolated points
plt.plot(x, y_pred, 'b-', label='Prediction')
plt.fill(np.concatenate([x, x[::-1]]), \
         np.concatenate([y_pred - 1.9600 * sigma,\
                        (y_pred + 1.9600 * sigma)[::-1]]),\
         alpha=.5, fc='b', ec='None', label='95% confidence interval')

plt.legend(loc='upper right')
plt.title('ptpos NNLO m = 171.5')
plt.xlabel('ptpos [GeV]')
plt.ylabel("Events")

plt.show()
```
